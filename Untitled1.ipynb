# **Dataset: Place365.**

```python
# Step 1: Download and extract Places365 test dataset
!wget http://data.csail.mit.edu/places/places365/test_256.tar
!tar -xvf test_256.tar

# Step 2: Import libraries
import os
import shutil
import random
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# Step 3: Create a new folder and pick 25 random images
os.makedirs("places25/images/classA", exist_ok=True)  # Dummy class folder

# Get all image filenames from test_256
all_images = [f for f in os.listdir("test_256") if f.endswith(".jpg")]

# Pick 25 random images
random_images = random.sample(all_images, 25)

# Copy them into the dummy class folder
for img in random_images:
    shutil.copy(f"test_256/{img}", f"places25/images/classA/{img}")

# Step 4: Define transformation (resize and convert to tensor)
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

# Step 5: Load dataset using ImageFolder
dataset = datasets.ImageFolder("places25/images", transform=transform)
loader = DataLoader(dataset, batch_size=25, shuffle=False)

# Step 6: Show 25 images in a 5x5 grid
images, labels = next(iter(loader))

plt.figure(figsize=(12, 12))
for i in range(25):
    plt.subplot(5, 5, i+1)
    img = images[i].permute(1, 2, 0)  # Change channel order for plotting
    plt.imshow(img)
    plt.title(f"Image {i+1}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
